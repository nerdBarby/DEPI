```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
file_path = 'E:/breast-cancer.csv'
data = pd.read_csv(file_path)
```

```python
# Encode the diagnosis (M -> 1, B -> 0)
data['diagnosis'] = data['diagnosis'].apply(lambda x: 'Malignant' if x == 'M' else 'Benign')

# Define columns for mean and worst values
mean_columns = [col for col in data.columns if 'mean' in col]
worst_columns = [col for col in data.columns if 'worst' in col]

# Separate data by diagnosis
benign_data = data[data['diagnosis'] == 'Benign']
malignant_data = data[data['diagnosis'] == 'Malignant']
```

```python
# Calculate means for both benign and malignant groups for mean and worst values
benign_means = benign_data[mean_columns].mean()
benign_worsts = benign_data[worst_columns].mean()

malignant_means = malignant_data[mean_columns].mean()
malignant_worsts = malignant_data[worst_columns].mean()
```

```python
# Create DataFrames for each comparison
comparison_df = pd.DataFrame({
    'Benign Mean': benign_means,
    'Benign Worst': benign_worsts,
    'Malignant Mean': malignant_means,
    'Malignant Worst': malignant_worsts
})
```

```python
# Improved color scheme
palette_benign = ['#1f77b4', '#ff7f0e']   # Blue, Orange
palette_malignant = ['#2ca02c', '#d62728']  # Green, Red
palette_mean = ['#9467bd', '#e377c2']      # Purple, Pink
palette_worst = ['#8c564b', '#17becf']     # Brown, Cyan
```

```python
# Plotting Benign Mean vs Benign Worst
plt.figure(figsize=(12, 6))
comparison_df[['Benign Mean', 'Benign Worst']].plot(kind='bar', figsize=(15, 6), color=palette_benign)
plt.title('Comparison of Benign Mean vs Benign Worst Feature Values')
plt.ylabel('Mean Feature Values')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show();
```

```python
# Plotting Malignant Mean vs Malignant Worst
plt.figure(figsize=(12, 6))
comparison_df[['Malignant Mean', 'Malignant Worst']].plot(kind='bar', figsize=(15, 6), color=palette_malignant)
plt.title('Comparison of Malignant Mean vs Malignant Worst Feature Values')
plt.ylabel('Mean Feature Values')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Plotting Benign Mean vs Malignant Mean
plt.figure(figsize=(12, 6))
comparison_df[['Benign Mean', 'Malignant Mean']].plot(kind='bar', figsize=(15, 6),color=palette_mean)
plt.title('Comparison of Benign Mean vs Malignant Mean Feature Values')
plt.ylabel('Mean Feature Values')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Plotting Benign Worst vs Malignant Worst
plt.figure(figsize=(12, 6))
comparison_df[['Benign Worst', 'Malignant Worst']].plot(kind='bar', figsize=(15, 6), color=palette_worst)
plt.title('Comparison of Benign Worst vs Malignant Worst Feature Values')
plt.ylabel('Mean Feature Values')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
